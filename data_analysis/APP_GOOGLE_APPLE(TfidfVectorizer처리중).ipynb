```python
from konlpy.tag import Okt
okt = Okt()
```

```python
stopwords = ['하다','있다','이다','되다','넘다','없다','자다']
# 토크나이징 함수 정의
# def tokenizer(raw, pos=["Noun","Alpha","Verb","Number"], stopword=stopwords):
def Tokenizer(raw, pos=["Noun","Verb","Modifier","Adjective"], stopword=stopwords):
    return [
        word for word, tag in okt.pos(
            raw,
            norm=True,   # normalize 그랰ㅋㅏ -> 그래ㅋㅋ
            stem=True    # stemming 바뀌나->바뀌다
            )
            if len(word) > 1 and tag in pos and word not in stopword
    ]
```

```python
# tokenizing
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=Tokenizer, max_df=0.70, min_df=2)
# vectorizer = TfidfVectorizer(tokenizer=tokenizer, use_idf=True)
features = vectorizer.fit_transform(df_app["user_comments"])
features.toarray()
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_model = LatentDirichletAllocation(n_components=5, random_state=111) #인스턴스화 #n_components 토픽의 갯수
lda_model.fit(features) #교육
```

```python
dictionary_list = vectorizer.get_feature_names_out()
```

```python
# concaterate/ zip은 1차원 행렬만 가능
list(zip(dictionary_list, lda_model.components_[0],lda_model.components_[1],lda_model.components_[2],lda_model.components_[3],lda_model.components_[4]))
```

```python
## 상위 단어 추출 
## 0 확률 1은 dictionary
topics_list = list()
for topic in lda_model.components_:
    df_datas = [topic, dictionary_list]
    df_topics = pd.DataFrame(data=df_datas)
    df_topics= df_topics.T
    df_topics = df_topics.sort_values(0, ascending=False)
    # print(df_topics[:3])
    topics_text = ' '.join(df_topics[1].values[:4])# 시리즈 형식으로 출력 get values from series / index 
    print(topics_text)
    topics_list.append(topics_text)
' '.join(df_topics[1].values[:4]) #시리즈를 하나에 문장 생성
topics_list_add = [['Topic0', 'Topic1', 'Topic2', 'Topic3', 'Topic4'],topics_list]
df_topics_keywords = pd.DataFrame(topics_list_add)
df_topics_keywords.T
```
