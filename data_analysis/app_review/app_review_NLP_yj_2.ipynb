```python
from mecab import MeCab
mecab = MeCab()
```

```python
mecab.pos(df_label_complete['review_clean_okt'][5])
```

```python
def mecab_tag(text):
    morphs_nouns = []
    morphs_nouns_list = []
    for sentence in text:
        # 명사, 형용사, 동사
        morphs = [morph for morph, tag in mecab.pos(sentence) if tag in ['NNG']]
        morphs_nouns.append(' '.join(morphs))
    
    return morphs_nouns
```

```python
morphs_nouns = mecab_tag(df_label_complete['review_clean_okt'])
```

```python
df_label_complete['review_clean_noun'] = morphs_nouns
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

### 부정

```python
# 텍스트 데이터 전처리 및 벡터화
vectorizer = TfidfVectorizer()
feature_negative = vectorizer.fit_transform(df_label_complete.query('label == 0')['review_clean_noun'])
```

```python
# LDA 모델 훈련
lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(feature_negative)
```

```python
# 훈련된 LDA 모델로 문서별 토픽 할당
document_topics = lda.transform(feature_negative)

# 각 문서의 주요 토픽 확인
for idx, topic in enumerate(document_topics):
    print(f"Document {idx} belongs to Topic {topic.argmax()}")
```

```python
# 토픽 모델링 결과 해석
feature_names = vectorizer.get_feature_names_out()
for topic_idx, topic in enumerate(lda.components_):
    # 토픽별로 상위 단어 10개 출력
    top_word_indices = topic.argsort()[:-10-1:-1]
    top_words = [feature_names[i] for i in top_word_indices if i < len(feature_names)]
    print(f"Topic {topic_idx}: {', '.join(top_words)}")
```

## 앱별로 긍, 부정 나누기

```python
tripadvisor = df_label_complete.query("app_name == 'tripadvisor'")
triple = df_label_complete.query("app_name == 'triple'")
myrealtrip = df_label_complete.query("app_name == 'myrealtrip'")
skyscanner = df_label_complete.query("app_name == 'skyscanner'")
```

### 트립어드바이저

#### 부정

```python
df_sheets = []
for sheet_name, df in tripadvisor.query("label == 0")['review_clean_noun'].items() :
  df_sheets.append(df)
```

```python
vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(df_sheets)
features.toarray()
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_model = LatentDirichletAllocation(n_components=3, random_state=111) #인스턴스화 #n_components 토픽의 갯수
lda_model.fit(features)
```

```python
dictionary_list = vectorizer.get_feature_names_out()
```

```python
## 상위 단어 추출 
## 0 확률 1은 dictionary
topics_list = list()
for topic in lda_model.components_:
    df_datas = [topic, dictionary_list]
    df_topics = pd.DataFrame(data=df_datas)
    df_topics= df_topics.T
    df_topics = df_topics.sort_values(0, ascending=False)
    # print(df_topics[:3])
    topics_text = ' '.join(df_topics[1].values[:4])# 시리즈 형식으로 출력 get values from series / index 
    print(topics_text)
    topics_list.append(topics_text)
' '.join(df_topics[1].values[:4]) #시리즈를 하나에 문장 생성
topics_list_add = [['Topic0', 'Topic1', 'Topic2'],topics_list]
df_topics_keywords = pd.DataFrame(topics_list_add)
df_topics_keywords.T
```

```python
topics_output = lda_model.transform(features)
```

```python
df_topics_score = pd.DataFrame(topics_output)
```

```python
df_topics_score['dominant_topic_number']=np.argmax(topics_output, axis=1)
```

```python
df_topics_score['sentences']=df_sheets
```

```python
df_topics_score.query("dominant_topic_number == 0")
```

## 결론

[부정]
- 부정 리뷰의 경우 대체로 앱 오류 문제가 많음.
- 사용이 불편하다 -> 사용자 입장에서 알기 쉽도록, 사용 편리하도록 페이지 디자인
- 예약 취소 수수료가 비싸다 -> 타 사이트들 보다 수수료를 낮춘다는 마케팅 가능
- 실 결제시 가격이 달라지는 문제 -> 우리 사이트는 처음 보여지는 가격과 결제시 가격이 같아요!! 라는 마케팅 가능

[긍정]
- 여러가지 상품의 가격을 비교해서 보는 기능
- 자주 이용하는 교통수단이나 자주 가는 지역 등등을 저장하는 기능
- 검색 옵션 많으면 좋다
- 다른 사용자와 일정을 공유하는 기능
