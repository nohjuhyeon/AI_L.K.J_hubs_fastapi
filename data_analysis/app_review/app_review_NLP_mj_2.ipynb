# 피클 불러오기

```python
app_pickle = 'model.pkl'

import pickle
with open(app_pickle,'rb') as pickle_file:
  app_pickle_classification_model = pickle.load(pickle_file)
app_pickle_classification_model
```

# 데이터 분류하기

```python
from sklearn.model_selection import train_test_split
df_app_train_train,df_app_train_test = train_test_split(df_app,test_size=0.3,random_state=42)
df_app['user_comments'].unique()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfVectorizer = TfidfVectorizer(use_idf=True,max_df=0.95,min_df=2,ngram_range=(1,2))
tfidfVectorizer.fit(df_app['user_comments'])
```

# 모델 선택하기

```python
train_feature = tfidfVectorizer.transform(df_app_train_train['user_comments']) # train
train_target = df_app_train_train['user_score'] # test
test_feature = tfidfVectorizer.transform(df_app_train_test['user_comments']) # train
test_target = df_app_train_test['user_score'] # test
```

```python
df_app_train_train
```

```python
app_pickle_classification_model.fit(train_feature,train_target)
```

```python
app_train_feature_predict = app_pickle_classification_model.predict(train_feature)
```

```python
from sklearn.metrics import classification_report
print(classification_report(train_target,app_train_feature_predict))
```

```python
app_test_feature_predict = app_pickle_classification_model.predict(test_feature)
```

```python
from sklearn.metrics import classification_report
print(classification_report(test_target,app_test_feature_predict))
```

```python
from sklearn.model_selection import GridSearchCV

# 탐색할 하이퍼 파라미터 지정
param_grid = {
    'n_estimators': [100],
    'max_features': ['sqrt'],
    'min_samples_split': [5],
    'min_samples_leaf': [2]
}

# 그리드 서치를 사용하여 최적의 파라미터 탐색
grid_search = GridSearchCV(estimator=app_pickle_classification_model, param_grid=param_grid, cv=5)
grid_search.fit(train_feature, train_target)
```

```python
# 최적의 파라미터 출력
print(grid_search.best_params_)
```

```python
# 최적의 모델로 예측
best_rf = grid_search.best_estimator_
a_test_feature_predict = best_rf.predict(test_feature)
```

```python
print(classification_report(test_target,a_test_feature_predict))
```

# 긍정 & 부정 확인

```python
from textblob import TextBlob

# 분석할 문장들의 리스트 = list_review

# 감정 분석 결과를 저장할 딕셔너리 초기화
sentiment_review = {'Positive': 0, 'Negative': 0, 'Neutral': 0}

# 각 문장에 대한 감정 분석
for sentence in list_review:
    sentence_str = str(sentence)
    blob = TextBlob(sentence_str)
    sentiment = blob.sentiment.polarity
    # 감정 평가를 바탕으로 빈도수 계산
    # if sentiment > 0:
    #     sentiment_review['Positive'] += 1
    # elif sentiment < 0:
    #     sentiment_review['Negative'] += 1
    # else:
    #     sentiment_review['Neutral'] += 1

    if sentiment > 0:
        sentiment_review['Neutral'] += 1
    elif sentiment < 0:
        sentiment_review['Negative'] += 1
    else:
        sentiment_review['Positive'] += 1


# 결과 출력
print(f"긍정적 문장의 수: {sentiment_review['Positive']}")
print(f"부정적 문장의 수: {sentiment_review['Negative']}")
print(f"중립적 문장의 수: {sentiment_review['Neutral']}")
```

```python
from textblob import TextBlob

# 각 문장에 대한 감정 분석 함수 정의
def analyze_sentiment(sentence, user_score):
    # user_score가 1이나 2인 경우 바로 부정으로 처리
    if user_score in ['1', '2']:
        return '부정'
    sentence_str = str(sentence)
    blob = TextBlob(sentence_str)
    sentiment = blob.sentiment.polarity
    # if sentiment > 0:
    #     return '긍정'
    # elif sentiment < 0:
    #     return '부정'
    # else:
    #     return '중립'

    if sentiment > 0:
        return '중립'
    elif sentiment < 0:
        return '부정'
    else:
        return '긍정'

# '감성분석' 컬럼에 감성 분석 값 넣기
# df_app['sentiment'] = df_app['user_comments'].apply(analyze_sentiment)
# apply 메서드에 lambda 함수를 사용하여 user_comments와 user_score를 함께 전달
df_app['sentiment'] = df_app.apply(lambda x: analyze_sentiment(x['user_comments'], x['user_score']), axis=1)
df_app.query("user_score == '3'")["sentiment"].value_counts()
```
