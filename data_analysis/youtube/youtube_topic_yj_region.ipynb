```python
from wordcloud import WordCloud
from collections import Counter

for i in range(len(df_youtube_comments)):
    region_tour=[]
    city_tour = []
    for j in region_gw:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('강원')
            city_tour.append(j)
    for j in region_gg:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('경기')
            city_tour.append(j)
    for j in region_gn:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('경남')
            city_tour.append(j)
    for j in region_gj:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('광주')   
            city_tour.append(j)
    for j in region_gb:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('경북')   
            city_tour.append(j)
    for j in region_sj:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('세종')   
            city_tour.append(j)
    for j in region_dg:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('대구')
            city_tour.append(j)
    for j in region_dj:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('대전')
            city_tour.append(j)
    for j in region_bs:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('부산')
            city_tour.append(j)
    for j in region_su:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('서울')
            city_tour.append(j)
    for j in region_us:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('울산')
            city_tour.append(j)
    for j in region_ic:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('인천')
            city_tour.append(j)
    for j in region_jn:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('전남')    
            city_tour.append(j)
    for j in region_jb:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('전북')
            city_tour.append(j)
    for j in region_jj:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('제주')
            city_tour.append(j)
    for j in region_cn:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('충남')
            city_tour.append(j)
    for j in region_cb:
        if j in df_youtube_comments['title_noun'][i]:
            region_tour.append('충북')
            city_tour.append(j)
    word_region = Counter(region_tour)
    most_common_region = word_region.most_common(1)
    most_common_region_keys = [word for word, count in most_common_region]
    word_city = Counter(city_tour)
    most_common_city = word_city.most_common(1)
    most_common_city_keys = [word for word, count in most_common_city]
    df_youtube_comments.loc[i, 'region'] = ", ".join(most_common_region_keys)
    df_youtube_comments.loc[i, 'city'] = ", ".join(city_tour)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

## 강원

```python
# 텍스트 데이터 전처리 및 벡터화
vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(df_youtube_comments.query('region == "강원"')['reply_clean_mecab_withoutregion'])
```

```python
lda = LatentDirichletAllocation(n_components=3, random_state=111)
lda.fit(features)
```

```python
dictionary_list = vectorizer.get_feature_names_out()
```

```python
## 상위 단어 추출 
## 0 확률 1은 dictionary
topics_list = list()
for topic in lda.components_:
    df_datas = [topic, dictionary_list]
    df_topics = pd.DataFrame(data=df_datas)
    df_topics= df_topics.T
    df_topics = df_topics.sort_values(0, ascending=False)
    # print(df_topics[:3])
    topics_text = ' '.join(df_topics[1].values[:4])# 시리즈 형식으로 출력 get values from series / index 
    print(topics_text)
    topics_list.append(topics_text)
' '.join(df_topics[1].values[:4]) #시리즈를 하나에 문장 생성
topics_list_add = [['Topic0', 'Topic1', 'Topic2'],topics_list]
df_topics_keywords = pd.DataFrame(topics_list_add)
df_topics_keywords.T
```

```python
df_sheets = []
for sheet_name, df in df_youtube_comments.query('region == "강원"')['reply_clean_mecab_withoutregion'].items() :
  df_sheets.append(df)
```

```python
text = ' '.join(map(str, df_sheets))
text_split = text.split()
text_split_count = Counter(text_split)
top_30 = pd.Series(text_split_count).sort_values(ascending=False)[:30]
wordCloud = WordCloud(background_color='white', font_path='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf')
wordCloud.generate_from_frequencies(top_30)

import matplotlib.pyplot as plt
plt.imshow(wordCloud)
plt.axis('off')
plt.show()
```
